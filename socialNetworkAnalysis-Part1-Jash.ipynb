1) create and visualize the symmetric and weighted networks of Bollywood actors.

```python
# Symmetric graph : A = B then B = A
Bollywood_symmetric = nx.Graph()
```

```python
Bollywood_symmetric.add_edge('Amitabh Bachchan', 'Abhishek Bachchan')
Bollywood_symmetric.add_edge('Amitabh Bachchan', 'Aamir Khan')
Bollywood_symmetric.add_edge('Amitabh Bachchan', 'Akshay Kumar')
Bollywood_symmetric.add_edge('Amitabh Bachchan', 'Dev Anand')
Bollywood_symmetric.add_edge('Abhishek Bachchan', 'Aamir Khan')
Bollywood_symmetric.add_edge('Abhishek Bachchan', 'Akshay Kumar')
Bollywood_symmetric.add_edge('Abhishek Bachchan', 'Dev Anand')
Bollywood_symmetric.add_edge('Dev Anand', 'Aamir Khan')
```

```python
plt.figure(figsize=(18,8))
nx.spring_layout(Bollywood_symmetric)
nx.draw_networkx(Bollywood_symmetric)
plt.show()
```

```python
# Weighted Graph
bollywood_weighted = nx.Graph()
bollywood_weighted.add_edge('Amitabh Bachchan','Abhishek Bachchan', weight=20)
bollywood_weighted.add_edge('Amitabh Bachchan','Aamir Khan', weight=5)
bollywood_weighted.add_edge('Amitabh Bachchan','Akshay Kumar', weight=7)
bollywood_weighted.add_edge('Amitabh Bachchan','Dev Anand', weight=5)
bollywood_weighted.add_edge('Abhishek Bachchan','Aamir Khan', weight=5)
bollywood_weighted.add_edge('Abhishek Bachchan','Akshay Kumar',weight=5)
bollywood_weighted.add_edge('Abhishek Bachchan','Dev Anand', weight=10)
bollywood_weighted.add_edge('Dev Anand','Aamir Khan',weight=2)
```

```python
plt.figure(figsize=(18,8))
nx.spring_layout(bollywood_weighted)
nx.draw_networkx(bollywood_weighted)
plt.show()
```

### As all the lines look the same, we cannot differentiate if two nodes have higher weights or lower weights. So let's plot the lines according to their weights. Solid lines between nodes mean they have an higher association in the network compared to other nodes.

```python
plt.figure(figsize=(12,6))
# let us set the cut-off point of the weight, here it is set as 8
elarge = [(u, v) for (u, v, d) in bollywood_weighted.edges(data=True) if d['weight'] > 8]
esmall = [(u, v) for (u, v, d) in bollywood_weighted.edges(data=True) if d['weight'] <=8]

# A popular position algorithm
pos = nx.spring_layout(bollywood_weighted)  

# nodes
nx.draw_networkx_nodes(bollywood_weighted, pos, node_size=700)

# define edges: 
#>8, regular line with more weights;
#<8, dashed thinner line
nx.draw_networkx_edges(bollywood_weighted, pos, edgelist=elarge,
                       width=6)
nx.draw_networkx_edges(bollywood_weighted, pos, edgelist=esmall,
                       width=3, alpha=0.5, edge_color='b', style='dashed')

# label for the nodes
nx.draw_networkx_labels(bollywood_weighted, pos, font_size=15, font_family='calibri')

plt.axis('off')
plt.show()
```
