# Social Network Analysis - Part 2

```python
import networkx as nx
import matplotlib.pyplot as plt
```

1) Visualize the network data

```python
#load the data
G_fb_net = nx.read_edgelist("facebook_combined.txt", create_using=nx.Graph(), nodetype=int)
```

```python
#network summary 
print(nx.info(G_fb_net))
```

```python
#visualize it
plt.figure(figsize = (18,8))
nx.draw_networkx (G_fb_net, with_labels=False, node_color = 'yellow')
plt.show()
```

```python
# A better visualization can be seen below
pos =  nx.spring_layout(G_fb_net)

# Vary node size by betweenness centrality
btwCent = nx.betweenness_centrality(G_fb_net, normalized = True, endpoints=True)
#a simple loop to define the size of each node based on centrality
node_size = [x * 10000 for x in btwCent.values()]
# Vary color by node degree
node_color = [20000.0 * G_fb_net.degree(x) for x in G_fb_net]

#Plot the network
plt.figure(figsize=(18,8))
nx.draw_networkx(G_fb_net, pos=pos, with_labels=False, node_color=node_color, node_size=node_size)
plt.axis('off')
```

2) Show the top five node labels with highest Between Centrality. 

```python
print("Top Five Node lables with highest Betweeness Centrality")
sorted(btwCent, key=btwCent.get, reverse=True)[:5]
```

3) Show the top five node labels with other two centrality measures:  Eigenvector Centrality and Degree Centrality

```python
# EigenVector Centrality
egn_vc = nx.eigenvector_centrality(G_fb_net)
print("Top Five Node labels - Eigenvector Centrality")
print(sorted(egn_vc, key=egn_vc.get, reverse=True)[:5])
# Degree Centrality
dc = nx.degree_centrality(G_fb_net)
print("\nTop Five Node labels - Degree Centrality")
print(sorted(dc, key=dc.get, reverse=True)[:5])
```
